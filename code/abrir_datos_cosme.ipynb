```python
curp = []
edo_civil = []
municipio = []
entidad = []
escolaridad = []
pareja = []
for item in range(len(s1_declaraciones)):
    curp.append(s1_declaraciones[item]['declaracion']['situacionPatrimonial']['datosGenerales']['curp'])
    edo_civil.append(s1_declaraciones[item]['declaracion']['situacionPatrimonial']['datosGenerales']['situacionPersonalEstadoCivil']['clave'])
    municipio.append(s1_declaraciones[item]['declaracion']['situacionPatrimonial']['domicilioDeclarante']['domicilioMexico']['municipioAlcaldia']['clave'])
    entidad.append(s1_declaraciones[item]['declaracion']['situacionPatrimonial']['domicilioDeclarante']['domicilioMexico']['entidadFederativa']['clave'])
    escolaridad.append(s1_declaraciones[item]['declaracion']['situacionPatrimonial']['datosCurricularesDeclarante']['escolaridad'])
    
    pareja.append(s1_declaraciones[item]['declaracion']['situacionPatrimonial'])

    
```

```python
s1_declaraciones_df = pd.DataFrame(zip(curp,edo_civil,municipio,entidad, escolaridad, pareja))
print(s1_declaraciones_df)
```
